### Chi-square test

#### Chi-Square Statistic:

* The Chi-Square statistic is a measure of how much the observed counts in the contingency table deviate from the counts we would expect if the two categorical variables (in this case, 'Category1' and another column) were independent.
* Higher values of the Chi-Square statistic indicate greater dissimilarity between observed and expected counts, suggesting a stronger association between the two categorical variables.
* By printing the Chi-Square statistic for each combination of 'Category1' with other columns, we can assess the strength of the relationship between 'Category1' and each of the other categorical variables individually.
#### P-value:

* The p-value associated with the Chi-Square statistic indicates the probability of observing a Chi-Square statistic as extreme as or more extreme than the one calculated, under the null hypothesis that the two categorical variables are independent.
* A low p-value (typically below a chosen significance level, such as 0.05) suggests that the observed association between the two categorical variables is unlikely to have occurred by chance alone.
* By printing the p-value for each combination of 'Category1' with other columns, we can determine whether the observed association is statistically significant.

In summary, printing the Chi-Square statistic and p-value for each combination of 'Category1' with other columns allows us to evaluate the strength and significance of the relationship between 'Category1' and each of the other categorical variables in the dataset.






```python
for column in df.columns[1:]:
    # Contingency table
    contingency_table = pd.crosstab(df['Growing_stress'], df[column])

    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Chi-Square Statistic for Category1 and {column}:\n", chi2)
    print(f"P-value for Category1 and {column}:", p)
```
